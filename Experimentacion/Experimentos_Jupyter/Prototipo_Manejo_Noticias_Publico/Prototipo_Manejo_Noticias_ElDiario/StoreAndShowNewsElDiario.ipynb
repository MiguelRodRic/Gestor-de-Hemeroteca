```python
# -*- coding: utf-8 -*-
"""
Created on Sun Feb 19 20:18:26 2017

@author: Miguel-Portatil
"""


#Library 
import urllib2
import pymongo
from contextlib import closing
from selenium.webdriver import Firefox
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from bs4 import BeautifulSoup
from pymongo import MongoClient


#set the url:
url = "http://www.eldiario.es/cultura"

#Create a request
request = urllib2.Request(url)

#Now, we can create a handle which we can read out later
handle = urllib2.urlopen(request)

#Read content
content = handle.read()

soup = BeautifulSoup(content, "lxml")

newsList = soup.find_all("article", {"class":"md-news-main"})

enlaces = [None]*len(newsList)
titulares = [None]*len(newsList)
autores = [None]*len(newsList)
for news in newsList:
    enlaceAux = news.find('a', href=True, title=True)
    enlaces[newsList.index(news)] = 'http://www.eldiario.es/' + enlaceAux['href']
    titulares[newsList.index(news)] = enlaceAux['title']
    autores[newsList.index(news)] = news.find('address').text.replace('\n','')

htmlNoticias = [None]*len(enlaces)
textoNoticias = [None]*len(enlaces)
fechaNoticias = [None]*len(enlaces)
comentariosNoticia = [None]*len(enlaces)

for enlace in enlaces:
    request = urllib2.Request(enlaces[enlaces.index(enlace)])
    handle = urllib2.urlopen(request)
    html = handle.read()
    htmlSoup = BeautifulSoup(html,"lxml")
    htmlNoticias[enlaces.index(enlace)] = htmlSoup.find_all("p", {"class":"mce"})
    fechaNoticias[enlaces.index(enlace)] = htmlSoup.find("div", {"class":"md-social"}).text.split('h',1)[0].replace('\n','')
    textoNoticias[enlaces.index(enlace)] = ""
    for parrafo in htmlNoticias[enlaces.index(enlace)]:
        textoNoticias[enlaces.index(enlace)] = textoNoticias[enlaces.index(enlace)] + parrafo.text
    #comentariosDiv = htmlSoup.find("div", {"class":"bb-comment-content-comments-ajax"})
    #comentariosAux = comentariosDiv.find("div",{"class":"comment-item-wrapper"})
    #comentariosNoticia[enlaces.index(enlace)] = comentariosDiv.find_all("blockquote")
#print comentariosNoticia     


print "OK"
```

```python
from selenium.webdriver import Firefox
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities

binary = FirefoxBinary(r'/usr/bin/firefox')
#System.setProperty("webdriver.gecko.driver","/usr/bin/wires")
caps = DesiredCapabilities.FIREFOX.copy()
caps["marionette"] = True;
#driver = Firefox(firefox_binary=binary, capabilities=caps, executable_path='/usr/bin/geckodriver')
sources = [None]*len(enlaces)
for enlace in enlaces:
    binary = FirefoxBinary(r'/usr/bin/firefox')
    #System.setProperty("webdriver.gecko.driver","/usr/bin/wires")
    caps = DesiredCapabilities.FIREFOX.copy()
    caps["marionette"] = True;
    with closing(Firefox(firefox_binary=binary, capabilities=caps, executable_path='/usr/bin/geckodriver')) as browser:
        browser.get(enlace)
        WebDriverWait(browser, timeout=10).until(lambda x: x.find_element_by_id('comments'))
        sources[enlaces.index(enlace)] = browser.page_source
    #print page_source
print 'OK'
```

```python
bloques = [None]*len(sources)
comments = [None]*len(sources)
for source in sources:
    soup = BeautifulSoup(sources[sources.index(source)],'lxml')
    bloques[sources.index(source)] = soup.find("div",{"class":"comment-holder pane"})
    comments[sources.index(source)] = bloques[sources.index(source)].find_all("div",{"class":"comment-item-wrapper2"})
    for comment in comments[sources.index(source)]:
        comments[sources.index(source)][comments[sources.index(source)].index(comment)]= comment.find("blockquote",{"class":"comment"}).text.replace('\n','').replace('\t','')
        #comment = comment.replace('\n','')
print 'OK'
```

```python
idInsertadas = [None]*len(enlaces)
cliente = MongoClient()
db = cliente.test_database
borradas = db.noticias.delete_many({})
noticias = db.noticias

for enlace in enlaces:
    indice = enlaces.index(enlace)
    autor = autores[indice]
    titular = titulares[indice]
    url = enlace
    fecha = fechaNoticias[indice]
    texto = textoNoticias[indice].encode('utf-8').decode('utf-8')
    noticia = { 'author': autor, 'title':titular,
               'link':url, 'publishDate':fecha, 'text':texto, 'source':'web', 'comentarios':comments[enlaces.index(enlace)]}
    idInsertadas[enlaces.index(enlace)]= noticias.insert_one(noticia)
print 'OK'
```
